```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
titanic = pd.read_csv('train.csv', index_col=0)
#ORTALAMA YAŞ
print(titanic.groupby(by="Sex").mean())
#30 YAŞ ALTI KURTULMA ORANI
print(titanic[(titanic['Age'] < 30)].mean())
#30 YAŞ ALTI ERKEK KURTULMA ORANI
print(titanic[(titanic['Age'] < 30) & (titanic['Sex'] =='male')].mean())
#30 YAŞ ALTI KADIN KURTULMA ORANI
print(titanic[(titanic['Age'] < 30) & (titanic['Sex'] =='female')].mean())

```
